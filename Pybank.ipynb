```python
import os
import csv

print("Financial Analysis")

budget_csv = os.path.join("Starter_Code","Starter_Code","PyBank","Resources","budget_data.csv")

months=[]
days=[]
total=0

profit_loss=0
profit_loss_change=[]
net_change=[]
net_total=0

avg_change=0
    
with open(budget_csv) as csvfile:
    csvreader=csv.reader(csvfile,delimiter=",")
    csv_header=next(csvreader)
    
    for row in csvreader:
        
        split_date = row[0].split("-")
        months.append(split_date[0])
        days.append(split_date[1])
        
        total+=(int(row[1]))


        profit_loss=int(row[1])
        profit_loss_change.append(profit_loss)
        
        net_total=sum(profit_loss_change)
        profit_loss_max_range = len(profit_loss_change)

    
    for profit_loss in range(1,profit_loss_max_range-1):
        x = profit_loss_change[profit_loss]-profit_loss_change[profit_loss-1]
        net_change.append(int(x))
    avg_change=(sum(net_change)/len(net_change))

    increase_value=0
    decrease_value=0
    
    increase_value = max(net_change)
    decrease_value = min(net_change)

    increase_change_index=net_change.index(increase_value) +1
    decrease_change_index=net_change.index(decrease_value) +1

    increase_date=months[(increase_change_index)]
    decrease_date=months[(decrease_change_index)]

    
    increase_day=days[(increase_change_index)]
    decrease_day=days[(decrease_change_index)]

    
    print("Total Months: "+str(len(months)))
    print("Total: " +"$"+ str(total)) 
    print("Average Change: $"+str(round(avg_change)))
    print("Greatest Increase of Profits: "+str(increase_date)+"-"+str(increase_day)+" $"+str(increase_value))
    print("Greatest Decrease of Profits: "+str(decrease_date)+"-"+str(decrease_day)+" $"+str(decrease_value))


output_file=os.path.join("Starter_Code","Starter_Code","PyBank","Financial_Analysis_Summary.txt")

with open(output_file,"w") as file:
    file.write("Financial Analysis")
    file.write("\n")
    file.write("\n")
    file.write("Total Months: " + str(len(months)))
    file.write("\n")
    file.write("Total"+"$"+str(total))
    file.write("\n")
    file.write("Average Change: "+"$"+str(avg_change))
    file.write("\n")
    file.write("Greatest Increase of Profits: "+str(increase_date)+"-"+str(increase_day)+" $"+str(increase_value))
    file.write("\n")
    file.write("Greatest Decrease of Profits: "+str(decrease_date)+"-"+str(decrease_day)+" $"+str(decrease_value))
    file.write("\n")
```
