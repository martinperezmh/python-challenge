```python
import csv
import os

poll_csv = os.path.join("PyPoll","Resources","election_data.csv")

total_votes=[]
total_num_votes=0
total_candidates=[]

votes_CCS=0
votes_DD=0
votes_RAD=0
votes = []

with open(poll_csv) as csvfile:
    csvreader= csv.reader(csvfile,delimiter=",")
    header = next(csvreader)
    
    for row in csvreader:
        total_votes.append(row[0])
        total_num_votes=len(total_votes)
        
        total_candidates=["Charles Casper Stockham","Diana DeGette","Raymon Anthony Doane"]
        if row[2]==total_candidates[0]:
            votes_CCS+=1
            votes_CCS_total=(int(votes_CCS))
        
        elif row[2]==total_candidates[1]:
            votes_DD+=1
            votes_DD_total=(int(votes_DD))
            
        elif row[2]==total_candidates[2]:
            votes_RAD+=1
            votes_RAD_total=(int(votes_RAD))
            
        votes=[votes_CCS_total,votes_DD,votes_RAD]    

            
    percentage_CCS=round((votes[0]/total_num_votes),5)*100
    percentage_DD=round((votes[1]/total_num_votes),5)*100
    percentage_RAD=round((votes[2]/total_num_votes),6)*100
    percentage_votes=[percentage_CCS,percentage_DD,percentage_RAD]
    
candidates_votes=list(zip(votes,total_candidates))

winner_votes_index=candidates_votes.index(max(candidates_votes))


print("Election Results")
print("Total Votes: "+str(total_num_votes))
print((total_candidates[0])+": "+ str(percentage_votes[0])+"%"+" ("+str(votes[0])+")")
print((total_candidates[1])+": "+ str(percentage_votes[1])+"%"+" ("+str(votes[1])+")")
print((total_candidates[2])+": "+ str(percentage_votes[2])+"%"+" ("+str(votes[2])+")")
print("Winner: "+(total_candidates[winner_votes_index]))

output_path=os.path.join("PyPoll","election_data_results.txt")

with open((output_path),'w') as file:
    file.write("Election Results")
    file.write("\n")
    file.write("\n")
    file.write("Total Votes: "+str(total_num_votes))
    file.write("\n")
    file.write("\n")
    file.write((total_candidates[0])+": "+ str(percentage_votes[0])+"%"+" ("+str(votes[0])+")")
    file.write("\n")
    file.write((total_candidates[1])+": "+ str(percentage_votes[1])+"%"+" ("+str(votes[1])+")")
    file.write("\n")
    file.write((total_candidates[2])+": "+ str(percentage_votes[2])+"%"+" ("+str(votes[2])+")")
    file.write("\n")
    file.write("\n")
    file.write("Winner: "+(total_candidates[winner_votes_index]))
    file.write("\n")
```
